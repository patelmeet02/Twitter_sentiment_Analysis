```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences
import pickle
import json
import numpy as np
import os

model_dir = r'C:\Users\User\OneDrive\Documents\Twitter_sentiment_analysis\sentiment_analysis_lstm_model'

# Load everything
model = load_model(os.path.join(model_dir, 'sentiment_lstm_model.keras'))

with open(os.path.join(model_dir, 'tokenizer.pickle'), 'rb') as f:
    tokenizer = pickle.load(f)

with open(os.path.join(model_dir, 'label_encoder.pickle'), 'rb') as f:
    le = pickle.load(f)

with open(os.path.join(model_dir, 'model_config.json'), 'r') as f:
    config = json.load(f)

max_words = config['max_words']
max_len = config['max_len']
embedding_dim = config['embedding_dim']

print("✓ Model loaded from directory!")
```

```python
test_text = "You are looking ugly"
sequence = tokenizer.texts_to_sequences([test_text])
padded = pad_sequences(sequence, maxlen=max_len, padding='post')
prediction = model.predict(padded, verbose=0)
predicted_label = le.inverse_transform([np.argmax(prediction)])[0]
print(f"Test prediction: '{test_text}' → {predicted_label}")
```
